```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('2017-fordgobike-tripdata.csv')
```

```python
list_2018=[]
for i in range(0,12):
    list_2018.append(pd.read_csv('2018{:02d}-fordgobike-tripdata.csv'.format(i+1)))
#df_2018_1=pd.read_csv('201801-fordgobike-tripdata.csv')
#df_2018_1=pd.read_csv('201801-fordgobike-tripdata.csv')
```

```python
list_2019=[]
for i in range(0,3):
    list_2019.append(pd.read_csv('2019{:02d}-fordgobike-tripdata.csv'.format(i+1)))
```

```python
df_2018 = pd.concat(list_2018)
df_2019 = pd.concat(list_2019)
```

```python
frame=[df,df_2018,df_2019]
df_master=pd.concat(frame,ignore_index=True,sort=True)
```

```python
from datetime import datetime
#format date column to work with datetime library
df_master['start_time']=df_master['start_time'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f'))
df_master['end_time']=df_master['end_time'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f'))
```

```python
#setup a df to have time as index
df_time_index_master=df_master.set_index('start_time')
```

```python
df_resample=df_time_index_master['duration_sec'].resample('D').count()
df_resample_week = df_resample.resample('W-WED').mean()
df_resample_day_and_week=pd.concat([df_resample,df_resample_week], axis=1)
df_resample_day_and_week.columns=['days','week']
#df_resample_day_and_week
df_resample_day_and_week['week']=df_resample_day_and_week['week'].interpolate()
ax=df_resample_day_and_week.plot(figsize=(15,10),title='Yearly bike trips per day trends')
ax.set_xlabel('Days')
ax.set_ylabel('Number of bike trips')
#do it hourly with mean
#do it weekly, join with daily for trend
```

```python
#seperate user types to see their indivdual trends
df_cus_time=df_time_index_master[df_time_index_master['user_type']=='Customer'].resample('D').count()
df_sub_time=df_time_index_master[df_time_index_master['user_type']=='Subscriber'].resample('D').count()
df_user_time=pd.concat([df_cus_time['duration_sec'],df_sub_time['end_time']], axis=1)
df_user_time.columns=['Costumers','Subscribers']
ax=df_user_time.plot(figsize=(15,10),title='Yearly bike trips per user type')
ax.set_xlabel('Days')
ax.set_ylabel('Number of bike trips')
```

```python
bike_id_array=df_master.groupby('bike_id').count().index
```

```python
for id_name in bike_id_array:
    print(id_name)
```

```python
spike_dic={}
for id_name in bike_id_array:
    spike_dic[id_name]=df_master[df_master['bike_id']==id_name]
```

```python
for id_name in bike_id_array:
    spike_dic[id_name]=spike_dic[id_name].sort_values('start_time',ascending=False)
```

```python
# end time is 7
# start time is 14
spike_dic[7082].iloc[0,7]
```

```python
for id_name in bike_id_array:
    for i in range(0,len(spike_dic[id_name])):
        if i == 0:
            spike_dic[id_name].loc[spike_dic[id_name].index[i], 'delta_time'] = None
        else:
            spike_dic[id_name].loc[spike_dic[id_name].index[i], 'delta_time'] = (spike_dic[id_name].iloc[i-1,7]-spike_dic[id_name].iloc[i,14]).total_seconds()
```

```python
spike_mean_dic={}
for id_name in bike_id_array:
    spike_mean_dic[id_name]=spike_dic[id_name]['delta_time'].mean()
```

```python
df_spike=pd.concat(spike_dic.values(), ignore_index=True)
```
